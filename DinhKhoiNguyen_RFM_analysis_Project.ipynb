####**2.3.2.Handling the target value**


After checking the dataset, there are some rules that i have notify and give a few assumptions for them:

  + For the orders having "**Quantity**" and "**Price**" but no "**CustomerID**", there is the assumption that these orders belongs to "IMPORT" section and don't need to be concern in our model
  + For the orders having no "**CustomerID**" and set "**Price**" ="*0*", there are some seasons have been figured out like "*lost, wrong imput*". For these data, we dont really need it.

```python
# Step 1: Drop rows with conditions
cond1 = records['CustomerID'].isna() & records['Quantity'].notna() & records['UnitPrice'].notna()
cond2 = records['CustomerID'].isna() & (records['UnitPrice'] == 0)
cond3 = records['UnitPrice'] == -11062.06

records_cleaned = records[~(cond1 | cond2 | cond3)].copy()
```

After processing the first missing data handling with above three conditions, all the problems have been first solved. Now, the dataset can be considered as clean for further steps.

```python
# Step 2: Create 'OrderPrice'
records['OrderPrice'] = records['Quantity'] * records['UnitPrice']

# Step 3: Remove customers with only negative OrderPrice
has_positive_order = records.groupby('CustomerID')['OrderPrice'].apply(lambda x: (x > 0).any())
valid_customers_order = has_positive_order[has_positive_order].index
invalid_customers_order = has_positive_order[~has_positive_order].index

records_valid = records[records['CustomerID'].isin(valid_customers_order)].copy()
Just_Returned_Order_Customer = records[records['CustomerID'].isin(invalid_customers_order)].copy()
```

###2.4.Modifying the Dataset for RFM Segmentation

#####2.4.1.Dropping unnecessary value

No need for these attributes anymore

```python
# Drop Quantity and UnitPrice first
records_valid = records_valid.drop(columns=['StockCode','Description'])

print(records_valid)
```

#####2.4.2.RFM component modification

######**A.Recency Score**

Taking 31/12/2011 as the markpoint for the calculation, the "**recency**" is formed by taking real time date between the lastest transaction date with the target date.

```python
# Make sure InvoiceDate is datetime type
records_valid['InvoiceDate'] = pd.to_datetime(records_valid['InvoiceDate'])

# Reference date to subtract from
reference_date = pd.to_datetime('2011-12-31')

# Get the most recent InvoiceDate per CustomerID
last_purchase = records_valid.groupby('CustomerID')['InvoiceDate'].max()

# Calculate recency as days difference
recency = (reference_date - last_purchase).dt.days

# Add recency column to your DataFrame (e.g., df_grouped)
records_valid['recency'] = records_valid['CustomerID'].map(recency)

print(records_valid[['CustomerID', 'recency']])
```

######**B.Frequency Score**


"**frequency**" shows the number of transactions shown during the active time of each customer.

```python
frequency = records_valid.groupby('CustomerID')['InvoiceNo'].nunique()
records_valid['frequency'] = records_valid['CustomerID'].map(frequency)

print(records_valid)
```

######C.Monetary Score

""**monetary**"" aka ""**OrderPrice**"" gives the value of the a total pay of each customer when combine every single transactions.

```python
# Drop 'Quantity' and 'UnitPrice' as they're no longer needed
records_valid = records_valid.drop(columns=['Quantity', 'UnitPrice'])

# Group by 'InvoiceNo' and aggregate relevant fields
records_valid = records_valid.groupby('InvoiceNo').agg({
    'OrderPrice': 'sum',
    'CustomerID': 'first',
    'Country': 'first',
    'recency': 'first',
    'frequency': 'first'
}).reset_index()

# Preview the resulting DataFrame
print(records_valid)
```

```python
# Group by CustomerID and aggregate
records_RFM = records_valid.groupby('CustomerID').agg({
    'OrderPrice': 'sum',
    'Country': 'first',
    'recency': 'first',
    'frequency': 'first'
}).reset_index().rename(columns={'OrderPrice': 'monetary'})
```

######D.Final Formatting Attributes

```python
#Remove duplicated edge values to retain only distinct boundaries.
recency_bins = np.unique(np.percentile(records_RFM['recency'], [0,20,40,60,80,100]))
frequency_bins = np.unique(np.percentile(records_RFM['frequency'], [0,20,40,60,80,100]))
monetary_bins = np.unique(np.percentile(records_RFM['monetary'], [0,20,40,60,80,100]))

#Apply pd.cut using these bin edges.
records_RFM['Recency_Score'] = pd.cut(records_RFM['recency'], bins=recency_bins, labels=range(len(recency_bins)-1,0,-1), include_lowest=True)
records_RFM['Frequency_Score'] = pd.cut(records_RFM['frequency'], bins=frequency_bins, labels=range(1,len(frequency_bins)), include_lowest=True)
records_RFM['Monetary_Score'] = pd.cut(records_RFM['monetary'], bins=monetary_bins, labels=range(1,len(monetary_bins)), include_lowest=True)

print(records_RFM)
```

```python
# Ensure all 3 score columns are string type before concatenation
records_RFM['RFM_Score'] = (
    records_RFM['Recency_Score'].astype(str) +
    records_RFM['Frequency_Score'].astype(str) +
    records_RFM['Monetary_Score'].astype(str)
)

# View result
print(records_RFM[['CustomerID', 'RFM_Score']].head())
```

```python
# Dictionary: RFM_Score → Segment
rfm_segments = {
    "Champions": ['555', '554', '544', '545', '454', '455', '445'],
    "Loyal": ['543', '444', '435', '355', '354', '345', '344', '335'],
    "Potential Loyalist": ['553', '551', '552', '541', '542', '533', '532', '531', '452', '451', '442', '441',
                           '431', '453', '433', '432', '423', '353', '352', '351', '342', '341', '333', '323'],
    "New Customers": ['512', '511', '422', '421', '412', '411', '311'],
    "Promising": ['525', '524', '523', '522', '521', '515', '514', '513', '425', '424', '413', '414', '415',
                  '315', '314', '313'],
    "Need Attention": ['535', '534', '443', '434', '343', '334', '325', '324'],
    "About To Sleep": ['331', '321', '312', '221', '213', '231', '241', '251'],
    "At Risk": ['255', '254', '245', '244', '253', '252', '243', '242', '235', '234', '225', '224', '153',
                '152', '145', '143', '142', '135', '134', '133', '125', '124'],
    "Cannot Lose Them": ['155', '154', '144', '214', '215', '115', '114', '113'],
    "Hibernating customers": ['332', '322', '233', '232', '223', '222', '132', '123', '122', '212', '211'],
    "Lost customers": ['111', '112', '121', '131', '141', '151']
}

# Flatten into a score → segment mapping
score_to_segment = {score: segment for segment, scores in rfm_segments.items() for score in scores}

# Apply mapping to a new column
records_RFM['Segment'] = records_RFM['RFM_Score'].map(score_to_segment)

# Optional: fill NaN as 'Others' or 'Unclassified'
records_RFM['Segment'] = records_RFM['Segment'].fillna('Unclassified')

# View result
print(records_RFM[['CustomerID','Country', 'RFM_Score', 'Segment']].head())
```

```python
# Tạo cột phân biệt UK và Non-UK
records_RFM['Region'] = records_RFM['Country'].apply(lambda x: 'UK' if x == 'United Kingdom' else 'Non-UK')

def summary_stats(df):
    print("=" * 60)
    print("                    ENHANCED SUMMARY STATISTICS")
    print("=" * 60)

    # ===================== PART 1: SUMMARY STATISTICS =====================
    print("\n🔍 PART 1: SUMMARY STATISTICS")
    print("-" * 50)

    # Overall stats
    total_customers = df['CustomerID'].nunique()
    total_revenue = df['monetary'].sum()
    avg_rfm = df[['recency', 'frequency', 'monetary']].mean()
    max_rfm = df[['recency', 'frequency', 'monetary']].max()
    min_rfm = df[['recency', 'frequency', 'monetary']].min()

    print(f"📊 Overall Statistics:")
    print(f"   Total Customers: {total_customers:,}")
    print(f"   Total Revenue: ${total_revenue:,.2f}")
    print(f"   Average RFM: R={avg_rfm['recency']:.1f}, F={avg_rfm['frequency']:.1f}, M=${avg_rfm['monetary']:.2f}")
    print(f"   Max RFM: R={max_rfm['recency']:.0f}, F={max_rfm['frequency']:.0f}, M=${max_rfm['monetary']:.2f}")
    print(f"   Min RFM: R={min_rfm['recency']:.0f}, F={min_rfm['frequency']:.0f}, M=${min_rfm['monetary']:.2f}")

    # Regional breakdown
    print(f"\n🌍 Regional Breakdown (UK vs Non-UK):")
    for region in ['UK', 'Non-UK']:
        region_df = df[df['Region'] == region]
        if len(region_df) > 0:
            customers = region_df['CustomerID'].nunique()
            revenue = region_df['monetary'].sum()
            avg_rfm_reg = region_df[['recency', 'frequency', 'monetary']].mean()
            max_rfm_reg = region_df[['recency', 'frequency', 'monetary']].max()
            min_rfm_reg = region_df[['recency', 'frequency', 'monetary']].min()

            print(f"\n   {region}:")
            print(f"      Customers: {customers:,} ({customers/total_customers*100:.1f}%)")
            print(f"      Revenue: ${revenue:,.2f} ({revenue/total_revenue*100:.1f}%)")
            print(f"      Avg RFM: R={avg_rfm_reg['recency']:.1f}, F={avg_rfm_reg['frequency']:.1f}, M=${avg_rfm_reg['monetary']:.2f}")
            print(f"      Max RFM: R={max_rfm_reg['recency']:.0f}, F={max_rfm_reg['frequency']:.0f}, M=${max_rfm_reg['monetary']:.2f}")
            print(f"      Min RFM: R={min_rfm_reg['recency']:.0f}, F={min_rfm_reg['monetary']:.0f}, M=${min_rfm_reg['monetary']:.2f}")


    # ===================== PART 2: TOP COUNTRIES ANALYSIS =====================
    print(f"\n\n🏆 PART 2: TOP NON-UK COUNTRIES ANALYSIS") # Corrected indentation here
    print("-" * 60)

    # Filter non-UK countries
    non_uk_df = df[df['Country'] != 'United Kingdom']

    if len(non_uk_df) > 0:
        # Top 5 by customer count
        top_customers = non_uk_df.groupby('Country').agg({
            'CustomerID': 'nunique',
            'monetary': 'sum',
            'frequency': 'sum'
        }).rename(columns={'CustomerID': 'Customers', 'monetary': 'Revenue', 'frequency': 'Total_Orders'})
        top_customers = top_customers.sort_values('Customers', ascending=False).head(5)

        # Top 5 by revenue
        top_revenue = non_uk_df.groupby('Country').agg({
            'CustomerID': 'nunique',
            'monetary': 'sum',
            'frequency': 'sum'
        }).rename(columns={'CustomerID': 'Customers', 'monetary': 'Revenue', 'frequency': 'Total_Orders'})
        top_revenue = top_revenue.sort_values('Revenue', ascending=False).head(5)

        # Top 5 by frequency (total orders)
        top_frequency = non_uk_df.groupby('Country').agg({
            'CustomerID': 'nunique',
            'monetary': 'sum',
            'frequency': 'sum'
        }).rename(columns={'CustomerID': 'Customers', 'monetary': 'Revenue', 'frequency': 'Total_Orders'})
        top_frequency = top_frequency.sort_values('Total_Orders', ascending=False).head(5)

        # Find countries in multiple lists
        countries_in_cust_rev = set(top_customers.index) & set(top_revenue.index)
        countries_in_cust_freq = set(top_customers.index) & set(top_frequency.index)
        countries_in_rev_freq = set(top_revenue.index) & set(top_frequency.index)
        countries_in_all_three = set(top_customers.index) & set(top_revenue.index) & set(top_frequency.index)

        # Function to get star markers for each country in context of current table
        def get_star_markers(country, current_table):
            # Count how many OTHER top 5 lists this country appears in
            appearances_in_others = 0
            if current_table != 'customers' and country in top_customers.index:
                appearances_in_others += 1
            if current_table != 'revenue' and country in top_revenue.index:
                appearances_in_others += 1
            if current_table != 'frequency' and country in top_frequency.index:
                appearances_in_others += 1

            # Total appearances including current table
            total_appearances = appearances_in_others + 1

            if total_appearances == 3:
                return " ⭐⭐⭐"
            elif total_appearances == 2:
                return " ⭐⭐"
            else:
                return ""

        print("🥇 Top 5 Countries by Customer Count:")
        print("   Rank | Country         | Customers | Revenue    | Orders")
        print("   " + "-" * 60)
        for i, (country, row) in enumerate(top_customers.iterrows(), 1):
            markers = get_star_markers(country, 'customers')
            print(f"   {i:2d}   | {country:<15} | {row['Customers']:6,} | ${row['Revenue']:8,.0f} | {row['Total_Orders']:6,.0f}{markers}")

        print(f"\n💰 Top 5 Countries by Revenue:")
        print("   Rank | Country         | Revenue    | Customers | Orders")
        print("   " + "-" * 60)
        for i, (country, row) in enumerate(top_revenue.iterrows(), 1):
            markers = get_star_markers(country, 'revenue')
            print(f"   {i:2d}   | {country:<15} | ${row['Revenue']:8,.0f} | {row['Customers']:6,} | {row['Total_Orders']:6,.0f}{markers}")

        print(f"\n🛒 Top 5 Countries by Total Orders (Frequency):")
        print("   Rank | Country         | Orders     | Customers | Revenue")
        print("   " + "-" * 60)
        for i, (country, row) in enumerate(top_frequency.iterrows(), 1):
            markers = get_star_markers(country, 'frequency')
            print(f"   {i:2d}   | {country:<15} | {row['Total_Orders']:8,.0f} | {row['Customers']:6,} | ${row['Revenue']:8,.0f}{markers}")

        # Summary of outstanding countries
        if countries_in_all_three:
            print(f"\n⭐⭐⭐ Countries excelling in ALL 3 categories:")
            for country in countries_in_all_three:
                cust_count = top_customers.loc[country, 'Customers'] if country in top_customers.index else top_revenue.loc[country, 'Customers'] if country in top_revenue.index else top_frequency.loc[country, 'Customers']
                revenue = top_customers.loc[country, 'Revenue'] if country in top_customers.index else top_revenue.loc[country, 'Revenue'] if country in top_revenue.index else top_frequency.loc[country, 'Revenue']
                orders = top_customers.loc[country, 'Total_Orders'] if country in top_customers.index else top_revenue.loc[country, 'Total_Orders'] if country in top_revenue.index else top_frequency.loc[country, 'Total_Orders']
                print(f"   • {country}: {cust_count:,} customers, ${revenue:,.0f} revenue, {orders:,.0f} orders")

        # Countries in exactly 2 categories
        countries_in_two = set()
        all_countries = set(top_customers.index) | set(top_revenue.index) | set(top_frequency.index)

        for country in all_countries:
            if country not in countries_in_all_three:
                appearances = 0
                categories = []
                if country in top_customers.index:
                    appearances += 1
                    categories.append("Customers")
                if country in top_revenue.index:
                    appearances += 1
                    categories.append("Revenue")
                if country in top_frequency.index:
                    appearances += 1
                    categories.append("Orders")

                if appearances == 2:
                    countries_in_two.add((country, tuple(categories)))

        if countries_in_two:
            print(f"\n⭐⭐ Countries excelling in 2 categories:")
            for country, categories in sorted(countries_in_two):
                print(f"   • {country}: Top 5 in {' & '.join(categories)}")
    else:
        print("   No non-UK countries found in the dataset.")

    print("\n" + "=" * 60)
# Call the function
summary_stats(records_RFM)
```

###3.4.Segment Distribution

```python
def segment_distribution(df):
    order = [
        'Champions', 'Loyal', 'Potential Loyalist', 'New Customers', 'Promising',
        'Need Attention', 'About To Sleep', 'At Risk', 'Cannot Lose Them',
        'Hibernating customers', 'Lost customers'
    ]

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    # 1. Overall bar plot (vertical)
    sns.countplot(data=df, x='Segment', order=order, ax=axes[0])
    axes[0].set_title('Segment Distribution - Overall')
    axes[0].tick_params(axis='x', rotation=45)  # Removed the invalid 'ha' argument
    axes[0].set_xlabel('Segment')
    axes[0].set_ylabel('Count')

    # Function to create horizontal bar plot with percentage for a region
    def plot_percentage_bar(region_df, ax, title):
        if len(region_df) == 0:
            ax.set_title(title)
            ax.text(0.5, 0.5, 'No data for this region',
                   horizontalalignment='center', verticalalignment='center',
                   transform=ax.transAxes, fontsize=12)
            ax.set_xticks([])
            ax.set_yticks([])
            return

        segment_counts = region_df['Segment'].value_counts()
        segment_percents = (segment_counts / len(region_df) * 100).reindex(order).fillna(0)

        # Only show segments that have data
        segment_percents_filtered = segment_percents[segment_percents > 0]

        if len(segment_percents_filtered) > 0:
            # Create horizontal bar plot
            bars = ax.barh(range(len(segment_percents_filtered)),
                          segment_percents_filtered.values,
                          color='skyblue')

            # Set labels and formatting
            ax.set_yticks(range(len(segment_percents_filtered)))
            ax.set_yticklabels(segment_percents_filtered.index)
            ax.set_xlabel('Percentage (%)')
            ax.set_title(title)
            ax.invert_yaxis()  # Highest on top

            # Add percentage labels on bars
            for i, (bar, value) in enumerate(zip(bars, segment_percents_filtered.values)):
                ax.text(bar.get_width() + 0.2, bar.get_y() + bar.get_height()/2,
                       f'{value:.1f}%', va='center', ha='left')

            # Set x-axis limit with some padding
            ax.set_xlim(0, segment_percents_filtered.max() * 1.15)
        else:
            ax.set_title(title)
            ax.text(0.5, 0.5, 'No data for this region',
                   horizontalalignment='center', verticalalignment='center',
                   transform=ax.transAxes, fontsize=12)
            ax.set_xticks([])
            ax.set_yticks([])

    # 2. UK bar chart - Check different possible UK region names
    uk_variations = ['United Kingdom', 'UK', 'U.K.', 'Britain', 'Great Britain']
    uk_df = df[df['Region'].isin(uk_variations)]

    # If no UK data found, try case-insensitive matching
    if len(uk_df) == 0:
        uk_df = df[df['Region'].str.contains('United Kingdom|UK|Britain', case=False, na=False)]

    plot_percentage_bar(uk_df, axes[1], 'Segment Distribution - UK')

    # 3. Other Regions bar chart
    other_df = df[~df['Region'].isin(uk_variations)]

    # If we used case-insensitive matching for UK, apply the same logic here
    if len(uk_df) > 0 and len(other_df) == len(df):
        # This means we found UK data with case-insensitive matching
        other_df = df[~df['Region'].str.contains('United Kingdom|UK|Britain', case=False, na=False)]

    plot_percentage_bar(other_df, axes[2], 'Segment Distribution - Other Regions')

    plt.tight_layout()
    plt.show()

    # Print some debugging information
    print(f"Total records: {len(df)}")
    print(f"UK records: {len(uk_df)}")
    print(f"Other regions records: {len(other_df)}")
    print(f"Unique regions in dataset: {df['Region'].unique()}")

# Call the function
segment_distribution(records_RFM)
```

*International Market Advantages:*

* Higher concentration of premium customers (Champions + Promising = 29.7% vs 23.9% in UK)
* Better customer engagement with fewer hibernating accounts
* Faster customer lifecycle progression to high-value segments

*UK Market Characteristics:*

* More traditional customer development patterns
* Higher proportion of customers at risk of churning
* Larger base of loyal but not premium customers
